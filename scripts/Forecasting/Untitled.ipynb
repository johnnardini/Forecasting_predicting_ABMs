```python
import glob, shutil
```

```python
find = "Pp_0.0_"
n = len(find)
for a in glob.glob("../../results/PDE_sims/*"):
    oldname = a
    
    location = a.find(find)
    
    if location != -1:
        newname = a[:location] + a[location +n:]

        shutil.copy(oldname,
                    newname)
```
